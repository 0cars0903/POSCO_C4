### 서울시의 구별 인구 data

```python
seoul = pd.read_csv('/home/piai/Bigdata 실습과제/1 Big Data 분석 기초/서울인구통계.txt',sep="\t", thousands=',', skiprows = 2, encoding='utf-8')
seoul_1 = seoul[['자치구', '계', '계.1', '계.2', '65세이상고령자']]
seoul_1 = seoul_1.drop([0])
dic = {'자치구' : '구별', '계' : '인구수', '계.1' : '한국인', '계.2' : '외국인', '65세이상고령자' : '고령자'}
seoul_1.rename(columns = dic, inplace=True)
seoul_1.head()
```

### 서울시의 구별 cctv 보유현황

```python
s_cctv = pd.read_excel('/home/piai/Bigdata 실습과제/1 Big Data 분석 기초/서울시CCTV.xlsx', names = ['기관명', '소계', '2011년 이전', '2012년', '2013년', '2014년', '2015년', '2016년', '2017년', '2018년', '2019년', '2020년', '2021년'], skiprows = 2)
s_cctv.rename(columns={s_cctv.columns[0] : '구별'}, inplace = True)
s_cctv.rename(columns={s_cctv.columns[1] : 'CCTV 수'}, inplace = True)
s_cctv.head()
```

```python
# 증가율 = 2016~2018 cctv 수/2015년 이전 cctv 수

s_cctv['최근증가율']=(s_cctv['2021년']+s_cctv['2020년']+s_cctv['2019년'])/(s_cctv['CCTV 수']-(s_cctv['2019년']+s_cctv['2020년']+s_cctv['2021년']))*100
s_cctv.sort_values(by='최근증가율',ascending=False).head(5)
s_cctv["구별"]=["강남구","강동구","강북구","강서구","관악구","광진구","구로구","금천구","노원구","도봉구",

"동대문구","동작구","마포구","서대문구","서초구","성동구","성북구","송파구","양천구","영등포구","용산구","은평구","종로구",

"중구","중랑구"]
```

```python
# 구별 ‘외국인비율＇과 ‘고령자비율＇을 계산
seoul_1['외국인 비율']=seoul_1['외국인']/seoul_1['인구수']*100

seoul_1['고령자 비율']=seoul_1['고령자']/seoul_1['인구수']*100

seoul_1.head()
```

## 데이터 수집 및 정제

```python
# 결측치 분석하기
df_raw = pd.read_csv('/home/piai/Bigdata 실습과제/1 Big Data 분석 기초/FITNESS_결측.csv', encoding='euc-kr')
df_raw.head()
```

```python
# 결측치 확인
df_raw.isnull().head() 
# 또는 df_raw.isna().head() 
df_raw.isnull().sum()
```

```python
# GENDER 변수의 결측치에 여성으로 대체
df_raw["GENDER"].fillna("여성", inplace = True)
```

```python
# GENDER 별 WEIGHT 평균 계산 및 대체
df_raw.groupby('GENDER')["WEIGHT"].agg("mean")
df_raw["WEIGHT"] = df_raw["WEIGHT"].fillna(df_raw.groupby('GENDER')["WEIGHT"].transform("mean").round(3))
df_raw.isnull().sum()
```

```python
# 이상치 확인 및 처리
df_raw.boxplot(figsize = (10,4))
# df_raw["RSTPULSE"] >= 100 # 28번째 data
df_raw.loc[27:29]
df_fitness = df_raw[df_raw['RSTPULSE'] < 100]
df_fitness.reset_index(drop = True, inplace = True)
df_fitness.isnull().sum(0)
```

```python
# scale 변환(방법 3가지)
# 숫자형 변수 선택
df_fitness_num = df_fitness.select_dtypes(exclude = "object")
# 방법 1 (평균과 표준편차)
df_scale_std = StandardScaler()
df_scale_std = df_scale_std.fit_transform(df_fitness_num)
df_scale_std = pd.DataFrame(df_scale_std, columns = df_fitness_num.columns)
# 방법 2 (최소-최대)
df_scale_minmax = MinMaxScaler().fit_transform(df_fitness_num)
df_scale_minmax = pd.DataFrame(df_scale_minmax, columns = df_fitness_num.columns)
# 방법 3 (로버스트)
df_scale_robust = RobustScaler().fit_transform(df_fitness_num)
df_scale_robust = pd.DataFrame(df_scale_robust, columns = df_fitness_num.columns)
```

```python
# scale 변환 결과 확인(RSTPUSRE 변수 비교)
df_rstpulse = pd.DataFrame()
df_rstpulse["Raw"] = df_fitness_num["RSTPULSE"]
df_rstpulse["Standard"] = df_scale_std["RSTPULSE"]
df_rstpulse["MinMax"] = df_scale_minmax["RSTPULSE"]
df_rstpulse["Robust"] = df_scale_robust["RSTPULSE"]
df_rstpulse.head().round(3) # data 변환 확인
df_rstpulse.describe().round(3)
hist_rstpulse = df_rstpulse[["Raw", "Standard", "MinMax", "Robust"]].hist(figsize=(10,5))
```
