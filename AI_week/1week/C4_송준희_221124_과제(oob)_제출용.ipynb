```python
class Set:
# 인스턴스 생성시 자동으로 호출되는 메서드
    def __init__(self, n = []):
        self.n = n
# 입력 값 중에서 중복값을 제거 후 리스트로 변환
        A = []
        for i in n:
            if i not in A:
                A.append(i)
        self.n = A

# 출력 구문
    def __str__(self):
        return "{"+"{0}".format(str(self.n)[1:-1])+"}"
    
# getitems
    def __getitem__(self, idx):
        return self.n[idx]
        
# Set의 길이
    def __len__(self):
        return len(self.n)

# 포함 되어 있는 원소 찾기
    def __contains__(self, x):
        if x in self.n:
            return True
        else:
            return False

# 포함되어 있지 않은 원소의 경우 추가
    def add(self, x):
        self.x = x
        if self.x not in self.n:
            self.n.append(self.x)
            
# 포함되어 있는 원소 제거
    def discard(self, x):
        self.x = x
        if self.x in self.n:
            self.n.remove(self.x)
            
# 초기화
    def clear(self):
        self.n = []

# 교집합
    def __and__(self, other):
        self.other = other
        # 리스트 생성
        N = [] # N
        O = [] # Other
        X = [] # 비교
        for i in range(len(self.n)):
            c1 = self.n[i]
            N.append(c1)
        for i in range(len(self.other)):
            c2 = self.other[i]
            O.append(c2)
        for i in N:
            if i in O:
                X.append(i)
        return Set(X)

# 차집합
    def __sub__(self, other):
        # 리스트 생성
        C = [] # 교집합
        D = [] # 차집합
        for i in range(len(self.n)):
            c1 = self.n[i]
            for j in range(len(other)):
                c2 = other[j]
                if c1 == c2:
                    C.append(c2)
            if c1 not in C:
                D.append(c1)
        return Set(D)
        
# 합집합
    def __or__(self, other):
        self.other = other
        B = []
        A = []
        for i in range(len(self.n)):
            c1 = self.n[i]
            B.append(c1)
        for j in range(len(self.other)):
            c2 = self.other[j]
            B.append(c2)
        for k in B:
            if k not in A:
                A.append(k)
        return Set(A)

# 부분집합
    def __le__(self, other): # <=
        self.other = other
        # 리스트 생성
        N = [] # N
        O = [] # Other
        X = [] # 비교
        for i in range(len(self.n)):
            c1 = self.n[i]
            N.append(c1)
        for i in range(len(self.other)):
            c2 = self.other[i]
            O.append(c2)
        for i in N:
            if i in O:
                X.append(i)
        if len(X) == len(N):
            return True
        else:
            return False

    def __ge__(self, other): # >=
        self.other = other
        # 리스트 생성
        N = [] # N
        O = [] # Other
        X = [] # 비교
        for i in range(len(self.n)):
            c1 = self.n[i]
            N.append(c1)
        for i in range(len(self.other)):
            c2 = self.other[i]
            O.append(c2)
        for i in O:
            if i in N:
                X.append(i)
        if len(X) == len(O):
            return True
        else:
            return False

# 합집합으로 변환(|=)
    def __ior__(self, other):
        self.other = other
        # 리스트 생성
        O = [] # Other
        X = [] # 비교
        for i in range(len(self.other)):
            c2 = self.other[i]
            O.append(c2)
        for i in O:
            self.add(i)
        return self

# 교집합으로 변환(&=)
    def __iand__(self, other):
        self.other = other
        # 리스트 생성
        N = [] # N
        O = [] # Other
        X = [] # 비교
        for i in range(len(self.n)):
            c1 = self.n[i]
            N.append(c1)
        for i in range(len(self.other)):
            c2 = self.other[i]
            O.append(c2)
        for i in N:
            if i in O:
                X.append(i)
        self.clear()
        for i in X:
            self.add(i)
        return self

# 차집합으로 변환(-=)
    def __isub__(self, other):
        # 리스트 생성
        C = [] # 교집합
        D = [] # 차집합
        for i in range(len(self.n)):
            c1 = self.n[i]
            for j in range(len(other)):
                c2 = other[j]
                if c1 == c2:
                    C.append(c2)
            if c1 not in C:
                D.append(c1)
        self.clear()
        for i in D:
            self.add(i)
        return self
```

```python
a = Set([1,2,3,4])
b = Set([1,2,3,4])

print(a)
print(b)
print()

a.discard(4)
b.discard(1)
print(a)
print(b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()

print(a | b)
print(a & b)
print(a - b)
print()

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print()

a = Set([1,2,3])
b = Set([3,4])
address_a = id(a)
a |= b
print(a)
print(address_a == id(a))
print()

a = Set([1,2,3])
b = Set([3,4])
address_a = id(a)
a &= b
print(a)
print(address_a == id(a))
print()

a = Set([1,2,3])
b = Set([3,4])
address_a = id(a)
a -= b
print(a)
print(address_a == id(a))
print()
```
